SIMPLE LINEAR REGRESSION

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline
```

```python
df = pd.read_csv("datasets/homeprices.csv")
df.head()
```

```python
# Plot a scatter plot to get an idea of the distribution of the data
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area, df.price, color="red")
```

```python
# Initialize your linear regression model object
reg = linear_model.LinearRegression()

#Fit the data/Train the data
reg.fit(df[['area']], df.price)
```

```python
# predict the price for a single value
area_to_predict = np.array([[3300]])
reg.predict(area_to_predict)
```

```python
# Linear regression calculates the values for coefficient and slope as shown below
print(reg.coef_)
print(reg.intercept_)
```

```python
# Plot line of best fit based on the linear regression model

plt.scatter(df.area, df.price, color="red")
plt.xlabel("area")
plt.ylabel("price")
plt.plot(df.area, reg.predict(df[["area"]]), color="green")
```

```python
# Predict price for values from a csv and save in output csv
d = pd.read_csv("datasets/areas.csv")
d.head()
```

```python
p = reg.predict(d[['area']])
d['prices'] = p
d.to_csv("datasets/predictions.csv", index=False)
```
